#Preparação captura dos dados

```python
import pandas as pd
#url_dados = 'https://github.com/alura-cursos/agendamento-hospitalar/blob/main/dados/A151346189_28_143_208.csv'
url_dados = '/content/A210653189_28_143_208.csv'
dados = pd.read_csv(url_dados, encoding= "ISO-8859-1", skiprows= 3, sep= ";", skipfooter=12,
                    thousands = ".", decimal = ",")

#Importando biblioteca pandas e carregando arquivo
#Tive problemas com o arquivo que foi carregado no github do Thiago
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

plt.style.use('ggplot')
ax = dados.plot(x="Unidade da Federação", y= "2021/Mar", kind = 'bar',figsize=(9,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}') )#Desafio 1 e dois
plt.title('Valor total gasto , no sistema SUS,  por estado em Março de 2021')
plt.show()

#Vizualização dos dados com tratamento para formatação do eixo Y
```

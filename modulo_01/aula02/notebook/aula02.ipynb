#Definição de bibliotecas e carga de dados

```python
import pandas as pd
dados_url="/content/A190306189_28_143_208.csv"
dados=pd.read_csv(dados_url, encoding= "ISO-8859-1", skiprows= 3, sep=";", skipfooter= 12, thousands=".", decimal=",")
```

```python
dados.head() #avaliando os 5 primeiros registros
```

```python
dados.tail() #mostrando as ultimas 5 linhas
```

```python
dados[["2021/Jan" , "2021/Fev" , "2021/Mar"]] #imprimindo slice do dataframe com as 3 ultimas colunas
```

```python
pd.options.display.float_format = '{:.2f}'.format
dados.mean() #ajustando formatacao de numero float
```

```python
#filtrando apenas a colunas que possuem dados (retirando colunas com observacoes nulas)
colunas_usaveis=dados.mean().index.tolist()
colunas_usaveis.insert(0, "Unidade da Federação") #inserindo a coluna Unidade da Federação para filtro na lista
dados[colunas_usaveis]
```

```python
dados[colunas_usaveis].head()
```

```python
usaveis = dados[colunas_usaveis]
usaveis = usaveis.set_index("Unidade da Federação") #alterando o index para utilizar a coluna Unidade da Federação
usaveis.head()
```

```python
usaveis[["2021/Jan" , "2021/Fev" , "2021/Mar"]] #retestando o comando slice com essa nova base
```

```python
usaveis.loc["35 São Paulo"] #filtrando a linha com as informações do Estado de São Paulo com comando .loc[]
```

```python
usaveis_sem_total=usaveis.drop(labels="Total", axis=1)
usaveis_sem_total.T.plot(figsize=(10,6)) #plotando um grafico, retirando coluna total para não atrapalhar a impressao
```

#Desafio #01: Reposicionar a legenda do gráfico: Dentro? Fora? Onde?

```python
import matplotlib.pyplot as plt

usaveis_sem_total=usaveis.drop(labels="Total", axis=1)
usaveis_sem_total.T.plot(figsize=(10,6))
plt.legend(bbox_to_anchor=(1.05,1)) #ajustando caixa de legenda para fora do gráfico
plt.show()

#parametros para o bbox_to_achor(a,b) (como funciona a tupla)
#--> a = o deslocamento a direita o grafico, se igual a 1 fica colado ao gráfico
#--> b = o deslocamento para cima ou para baixo do gráfico, onde
#-->       0 fica abaixo do gráfico
#-->       1 fica ao lado do gráfico
#-->       2 fica acima do gráfico
```
